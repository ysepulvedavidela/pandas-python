El metodo value_counts(), es de los mas utiles de pandas, que muestra los valores unicos de tu DataFrame.

Hay dos cosas importantes con respecto a value_counts:
La primera es que al ser valores unicos el primer valor es el valor más frencuente, el segundo el más frecuente, y asi sucesivamente, este orden se puede alterar cambiando el parametro Ascending que quiere decir ascendente, y ponerlo en verdadero.
Lo segundo es que dropna es un parametro que permite que si esta en verdadero descarta los valores nulos, pero si se pone en falso, no descarta los valores nulos, esto sirve cuando tenemos un numero considerable de valores nulos los cuales podemos analizar.

```python
import pandas as pd
```

```python
tabla = pd.read_csv('C:/Users/Yeremy/Documents/Proyectos/ventas-pandas/ventasTotales.csv', skiprows=5, sep=';')
```

```python
tabla.head()
```

```python
tabla.Vendedor.value_counts()
```

```python
tabla.Región.value_counts()
```

```python
tabla.Región.value_counts(ascending=True)
```

```python
tabla.Región.value_counts(ascending=True, dropna=False)
```

Para ordenar valores existe el metodo sort_values(), la cual ordena valores de una serie de datos o de un dataframe

```python
tabla.Vendedor.sort_values()
```

```python
tabla.sort_values(by=['Vendedor','Ganancia'])
```

La Condiciones Booleanas, son una manera de indexar nuestros datos según determinadas condiciones que permiten filtrar la información de nuestros DataFrames a traves de la condicion de los registros que pedimos, si la condición es verdadera o falsa, se van a seleccionar aquellas filas que cumplan dicha condición.
Si se desea controlar mas de una condición o un vector booleano a la vez, estas se deben agrupar entre parentesis o corchetes para que las operaciones se lleven a cabo correctamente.

```python
tabla.Artículo == 'Gorras'
```

```python
tabla[tabla.Artículo == 'Gorras']
```

```python
tabla[(tabla.Artículo == 'Gorras') & (tabla.Región == 'Norte')]
```

Para la busqueda de objetos en nuestros DataFrames, como por ejemplo, en la busqueda para saber si contiene una expresión, una frase o una letra inclusive o si es numerica o no. Disponemos del atributo str para usar con todas las series de nuestro DataFrame y un atributo que permite acceder a diferentes metodos como contains, startswidth o isnumeric

```python
tabla.Vendedor.str.contains('Trevor')
```

```python
tabla[tabla.Vendedor.str.contains('Trevor')]
```

```python
tabla[tabla.Vendedor.str.startswith('T')]
```
